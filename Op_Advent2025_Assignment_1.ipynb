## Question 1: UBOS Multi-District Population & Growth Forecast

```python

import numpy as np
from statistics import mean, median, variance, stdev
import matplotlib.pyplot as plt

# Population data
kampala = np.array([1200, 1250, 1300, 1350, 1420, 1500, 1580, 1650, 1720, 1800])
wakiso = np.array([950, 1000, 1070, 1150, 1220, 1300, 1390, 1480, 1570, 1670])
gulu = np.array([320, 330, 345, 360, 375, 390, 410, 430, 455, 480])

# Compute statistics
kampala_stats = {
    "Mean": mean(kampala.tolist()),
    "Median": median(kampala.tolist()),
    "Variance": variance(kampala.tolist()),
    "Standard Deviation": stdev(kampala.tolist())
}
wakiso_stats = {
    "Mean": mean(wakiso.tolist()),
    "Median": median(wakiso.tolist()),
    "Variance": variance(wakiso.tolist()),
    "Standard Deviation": stdev(wakiso.tolist())
}
gulu_stats = {
    "Mean": mean(gulu.tolist()),
    "Median": median(gulu.tolist()),
    "Variance": variance(gulu.tolist()),
    "Standard Deviation": stdev(gulu.tolist())
}

print("Kampala Stats:", kampala_stats)
print("Wakiso Stats:", wakiso_stats)
print("Gulu Stats:", gulu_stats)

# Fibonacci sequence
def fibonacci(n):
    seq = [0, 1]
    while len(seq) < n:
        seq.append(seq[-1] + seq[-2])
    return seq[:n]

fib_growth = fibonacci(5)

# Projected growth
kampala_proj = kampala[-1] + np.array(fib_growth)
wakiso_proj = wakiso[-1] + np.array(fib_growth)
gulu_proj = gulu[-1] + np.array(fib_growth)

# Variance comparison
print("Variance Comparison - Kampala:", {
    "Actual": variance(kampala.tolist()),
    "Projected": variance(kampala_proj.tolist())
})
print("Variance Comparison - Wakiso:", {
    "Actual": variance(wakiso.tolist()),
    "Projected": variance(wakiso_proj.tolist())
})
print("Variance Comparison - Gulu:", {
    "Actual": variance(gulu.tolist()),
    "Projected": variance(gulu_proj.tolist())
})

# Plot
years_actual = list(range(1, 11))
years_proj = list(range(11, 16))
plt.figure(figsize=(10, 6))
plt.plot(years_actual, kampala, label="Kampala Actual", marker='o')
plt.plot(years_proj, kampala_proj, label="Kampala Projected", marker='o')
plt.plot(years_actual, wakiso, label="Wakiso Actual", marker='s')
plt.plot(years_proj, wakiso_proj, label="Wakiso Projected", marker='s')
plt.plot(years_actual, gulu, label="Gulu Actual", marker='^')
plt.plot(years_proj, gulu_proj, label="Gulu Projected", marker='^')
plt.title("UBOS Multi-District Population Forecast")
plt.xlabel("Year")
plt.ylabel("Population (in thousands)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Question 2: Solar Micro-Grid Simulation

```python

from scipy.linalg import solve
import numpy as np
from statistics import mean, variance, stdev
import matplotlib.pyplot as plt

# Simulated daily demands
demands = [(100, 80), (110, 90), (120, 100), (130, 110), (140, 120), (150, 130), (160, 140)]
results = []

for D1, D2 in demands:
    A = np.array([[3, 2], [4, 1]])
    b = np.array([D1, D2])
    x, y = solve(A, b)
    results.append((x, y))

results = np.array(results)
solar = results[:, 0]
battery = results[:, 1]

print("Solar Usage Stats:", {
    "Mean": mean(solar.tolist()),
    "Variance": variance(solar.tolist()),
    "Std Dev": stdev(solar.tolist())
})
print("Battery Usage Stats:", {
    "Mean": mean(battery.tolist()),
    "Variance": variance(battery.tolist()),
    "Std Dev": stdev(battery.tolist())
})

plt.figure(figsize=(8, 5))
plt.plot(range(1, 8), solar, label="Solar Panels", marker='o')
plt.plot(range(1, 8), battery, label="Batteries", marker='s')
plt.title("Daily Solar vs Battery Usage")
plt.xlabel("Day")
plt.ylabel("Energy Units")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Question 3: Lake Victoria Fish Export Risk Model

```python

from statistics import mean, median, variance
import matplotlib.pyplot as plt

# Fibonacci stock
def fibonacci(n):
    seq = [0, 1]
    while len(seq) < n:
        seq.append(seq[-1] + seq[-2])
    return seq[:n]

stock = fibonacci(15)
prices = [12000, 12500, 11800, 13000, 12800, 12200, 12600, 12400, 12900, 12700, 12300, 12100, 11900, 13100, 12050]
revenue = [s * p for s, p in zip(stock, prices)]

print("Revenue Stats:", {
    "Mean": mean(revenue),
    "Median": median(revenue),
    "Variance": variance(revenue)
})

risk = "High risk" if variance(revenue) > 50000 else "Low risk"
print("Risk Level:", risk)

plt.figure(figsize=(8, 5))
plt.plot(stock, revenue, marker='o')
plt.title("Fish Stock vs Revenue")
plt.xlabel("Fish Stock")
plt.ylabel("Revenue (UGX)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Question 4: Weather & Agriculture Analysis

```python

import numpy as np
import matplotlib.pyplot as plt

kampala_rain = np.array([120, 140, 180, 200, 220, 180, 90, 70, 60, 100, 110, 130])
mean_rain = mean(kampala_rain.tolist())
classification = ["Good for maize" if r > 150 else "Drought risk" for r in kampala_rain]
print("Rainfall Classification:", classification)

gulu_rain = np.array([100, 130, 160, 190, 210, 170, 80, 60, 50, 90, 100, 120])
cos_sim = np.dot(kampala_rain, gulu_rain) / (np.linalg.norm(kampala_rain) * np.linalg.norm(gulu_rain))
print("Cosine Similarity:", cos_sim)

plt.figure(figsize=(8, 5))
plt.plot(range(1, 13), kampala_rain, label="Kampala", marker='o')
plt.plot(range(1, 13), gulu_rain, label="Gulu", marker='s')
plt.title("Monthly Rainfall Comparison")
plt.xlabel("Month")
plt.ylabel("Rainfall (mm)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Question 5: Taxi Transport Revenue & Variability

```python

import numpy as np
from statistics import mean, variance, stdev
from scipy.linalg import solve
import matplotlib.pyplot as plt

passengers = np.array([35, 40, 42, 50, 55, 60, 48, 52, 47, 45])
revenue = passengers * 2000

print("Revenue Stats:", {
    "Mean": mean(revenue.tolist()),
    "Variance": variance(revenue.tolist()),
    "Std Dev": stdev(revenue.tolist())
})

# Supply-demand system
A = np.array([[2, 3], [1, 2]])
b = np.array([100, 70])
x, y = solve(A, b)
print("Supply-Demand Solution:", {"Supply": x, "Demand": y})

# Forecast
forecast = mean(revenue[-3:].tolist())
print("Forecasted Revenue (Day 11):", forecast)

plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), revenue, label="Actual Revenue", marker='o')
plt.plot(11, forecast, label="Forecasted Revenue", marker='x', color='red')
plt.title("Taxi Revenue Forecast")
plt.xlabel("Day")
plt.ylabel("Revenue (UGX)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
